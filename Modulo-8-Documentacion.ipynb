```python
# Creacion de un diccionario
planet = {
    'name': 'Earth',
    'moons': 1
}

# Para imprimir variables se usa el metodo GET (Regresa None en caso de no existir la clave)
print(planet.get('name'))

# planet['name'] es idéntico a usar planet.get('name') (Arroja un KeyError en caso de no existir la clave)
print(planet['name'])

# Modificar variable
planet.update({'name': 'Makemake'})
print(planet.get('name'))

# Otro metodo de modificar variable
planet['name'] = 'Makemake'
print(planet['name'])

# Diferencia entre metodos de actualizacion
# Usando update
planet.update({
    'name': 'Jupiter',
    'moons': 79
})

# Usando corchetes
planet['name'] = 'Jupiter'
planet['moons'] = 79

# Para agregar una nueva clave basta con declararla
planet['orbital period'] = 4333

# el diccionario planet ahora contiene: {
#   name: 'jupiter'
#   moons: 79
#   orbital period: 4333
# }

# Para eliminar una clave en necesario usar la funcion pop
planet.pop('orbital period')

# el diccionario planet ahora contiene: {
#   name: 'jupiter'
#   moons: 79
# }
```

```python
# Se puede anidar los datos en mas diccionarios
planet['diameter (km)'] = {
    'polar': 133709,
    'equatorial': 142984
}

# el diccionario planet ahora contiene: {
#   name: 'Jupiter'
#   moons: 79
#   diameter (km): {
#      polar: 133709
#      equatorial: 142984
#   }
# }

print({planet['name']}, 'polar diameter:', {planet['diameter (km)']['polar']})

# Salida: Jupiter polar diameter: 133709
```

```python
rainfall = {
    'october': 3.5,
    'november': 4.2,
    'december': 2.1
}

for key in rainfall.keys():
    print(f'{key}: {rainfall[key]}cm')

# Salida:
# october: 3.5cm
# november: 4.2cm
# december: 2.1cm

# Validar si existe una clave en lista
# El valor de december: 2.1cm

# Si, 'december' existe en diccionario
if 'december' in rainfall:
    # rainfall [en la posición december] es igual a
    # rainfall [en la posición december] + 1 (2.1+1)
    rainfall['december'] = rainfall['december'] + 1

# Si no:
else:

    # rainfall [en la posición december] es igual a 1
    rainfall['december'] = 1
print('Cambio de diciembre:',rainfall['december'])
# Como december si existe, el valor será 3.1

# Suma de todos los datos
#Total de precipitaciones 0
total_rainfall = 0

# Para cada valor en los valores de rainfall
for value in rainfall.values():
    
    # El total de las precipitaciones será igual a ese mismo + el valor que se está iterando

    total_rainfall = total_rainfall + value

# Muestra 'Hay un total de precipitaciones (el valor total) en centímetros en el último cuarto (haciendo referencia al cuarto del año)

print(f'There was {total_rainfall}cm in the last quarter')

# Salida:
# There was 10.8cm in the last quarter
```
